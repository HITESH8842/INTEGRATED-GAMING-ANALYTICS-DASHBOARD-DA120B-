```python
# RQ1: Price vs Player Engagement

import pandas as pd
import matplotlib.pyplot as plt
import os

os.makedirs("Figures", exist_ok=True)
os.makedirs("Tables", exist_ok=True)

rq1 = games_df.copy()

needed_cols = ['price', 'average_playtime_forever']

rq1 = rq1[needed_cols].dropna()

rq1['price'] = pd.to_numeric(rq1['price'], errors='coerce')
rq1['average_playtime_forever'] = pd.to_numeric(
    rq1['average_playtime_forever'],
    errors='coerce'
)

rq1 = rq1.dropna()

rq1['price_group'] = pd.cut(
    rq1['price'],
    bins=[-1, 0, 10, 30, 60, 1000],
    labels=['Free', 'Low', 'Medium', 'High', 'Very High']
)

rq1_table = (
   rq1.groupby(
    'price_group',
    observed=False
)['average_playtime_forever']
    .mean()
    .reset_index()
)

rq1_table.to_csv("Tables/RQ1_price_engagement.csv", index=False)

display(rq1_table)

plt.figure(figsize=(8,5))
plt.bar(
    rq1_table['price_group'].astype(str),
    rq1_table['average_playtime_forever']
)
plt.title("RQ1: Game Price and Player Engagement")
plt.xlabel("Price Group")
plt.ylabel("Average Playtime Forever")
plt.tight_layout()
plt.savefig("Figures/RQ1_price_engagement.png")
plt.show()
```
