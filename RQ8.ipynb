```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

os.makedirs("Figures", exist_ok=True)
os.makedirs("Tables", exist_ok=True)

# -----------------------------
# Load datasets
# -----------------------------
games_df = pd.read_csv("Steam Games Dataset/games.csv", low_memory=False)
vgsales_df = pd.read_csv("Video Game Sales/vgsales.csv", low_memory=False)

# Auto-load Steam Store dataset
steam_files = glob.glob("Steam Store Games (Clean dataset)/*.csv")

if len(steam_files) > 0:
    steam_store_df = pd.read_csv(steam_files[0], low_memory=False)
    print("Steam Store file loaded:", steam_files[0])
else:
    steam_store_df = pd.DataFrame()
    print("Steam Store CSV not found.")

# -----------------------------
# Standardize column names
# -----------------------------
for df in [games_df, vgsales_df, steam_store_df]:
    if not df.empty:
        df.columns = (
            df.columns
            .str.strip()
            .str.lower()
            .str.replace(" ", "_")
            .str.replace("-", "_")
        )

print("Games Dataset:", games_df.shape)
print("Video Game Sales Dataset:", vgsales_df.shape)
print("Steam Store Dataset:", steam_store_df.shape)

display(games_df.head())
display(vgsales_df.head())
display(steam_store_df.head())

# ==========================================================
# RQ8 TABLE (Your Original Code - No Changes)
# ==========================================================

rq8 = games_df[['release_date', 'peak_ccu', 'average_playtime_forever']].copy()

rq8['release_date'] = pd.to_datetime(
    rq8['release_date'],
    errors='coerce',
    dayfirst=False
)

rq8['release_year'] = rq8['release_date'].dt.year

rq8['peak_ccu'] = pd.to_numeric(
    rq8['peak_ccu'],
    errors='coerce'
)

rq8['average_playtime_forever'] = pd.to_numeric(
    rq8['average_playtime_forever'],
    errors='coerce'
)

rq8 = rq8.dropna()

rq8_table = (
    rq8.groupby('release_year')[['peak_ccu', 'average_playtime_forever']]
    .mean()
    .reset_index()
)

display(rq8_table.head())

rq8_table.to_csv(
    "Tables/RQ8_Game_Age_vs_Player_Activity.csv",
    index=False
)

# ==========================================================
# RQ8 FIGURE (Added Only)
# ==========================================================

# In your dataset the release date is actually stored in 'name'
# and Peak CCU is stored in 'estimated_owners'

rq8_plot = games_df[['name', 'estimated_owners']].copy()

rq8_plot.columns = [
    'release_date',
    'peak_ccu'
]

# Convert release date
rq8_plot['release_date'] = pd.to_datetime(
    rq8_plot['release_date'],
    format='%b %d, %Y',
    errors='coerce'
)

# Extract year
rq8_plot['release_year'] = rq8_plot['release_date'].dt.year

# Convert Peak CCU
rq8_plot['peak_ccu'] = pd.to_numeric(
    rq8_plot['peak_ccu'],
    errors='coerce'
)

# Remove missing values
rq8_plot = rq8_plot.dropna()

# Average Peak CCU by release year
rq8_plot = (
    rq8_plot.groupby('release_year')['peak_ccu']
    .mean()
    .reset_index()
)

# Plot
plt.figure(figsize=(10,5))

plt.plot(
    rq8_plot['release_year'],
    rq8_plot['peak_ccu'],
    marker='o',
    linewidth=2
)

plt.title("RQ8: Game Age vs Player Activity")
plt.xlabel("Release Year")
plt.ylabel("Average Peak CCU")
plt.grid(True)

plt.tight_layout()

plt.savefig(
    "Figures/RQ8_Game_Age_vs_Player_Activity.png",
    dpi=300
)

plt.show()
```
